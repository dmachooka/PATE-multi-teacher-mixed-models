```python
import seaborn as sns
import matplotlib.pyplot as plt

# Aesthetic setup
sns.set_theme(style="whitegrid", context="talk", font_scale=1.15)
sns.set_palette("Paired")  # ggplot-inspired color scheme (thanks)
plt.figure(figsize=(18, 9))
ax = sns.barplot(
    data=df,
    x="Dataset",
    y="Accuracy",
    hue="Architecture",
    errorbar=None
)

# Title and labels
plt.title("Model Accuracy Comparison Across Datasets", fontsize=20, weight='bold', pad=20)
plt.xlabel("Dataset", fontsize=14)
plt.ylabel("Accuracy (%)", fontsize=14)

# Smaller legend with tighter spacing ( adjust as need)
legend = plt.legend(
    title="Architecture",
    bbox_to_anchor=(1.02, 1),
    loc='upper left',
    frameon=True,
    fontsize=9,
    title_fontsize=10,
    labelspacing=0.3,
    borderpad=0.3
)

# Clean aesthetics
sns.despine()
plt.tight_layout()
plt.show()
```

```python
g = sns.FacetGrid(df, col="Dataset", hue="Architecture", col_wrap=2, height=4.5, aspect=1.2, palette="Paired")
g.map(sns.scatterplot, "Data_dependent_epsilon", "Accuracy", s=100, edgecolor="black")
g.set(xscale="log")
g.add_legend()
g.set_axis_labels("Data-dependent  (log scale)", "Accuracy (%)")
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle("PrivacyUtility Relationship Across Datasets", fontsize=18, weight='bold')
sns.despine()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid", context="talk", font_scale=1.1)
sns.set_palette("Set2")

sns.regplot(
    data=df,
    x="Data_dependent_epsilon",
    y="Accuracy",
    scatter_kws={'s': 80, 'alpha': 0.7},
    line_kws={'color': 'black', 'lw': 2},
    logx=True
)

plt.title("Effect of Privacy () on Accuracy", fontsize=18, weight='bold', pad=15)
plt.xlabel("Data-dependent  (log scale)")
plt.ylabel("Accuracy (%)")
sns.despine()
plt.tight_layout()
plt.show()
```
